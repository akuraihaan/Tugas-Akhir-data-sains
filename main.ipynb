### Mengimport Library python

```python
import os
import re
from collections import Counter
import matplotlib.pyplot as plt
import string
```

### Membersihkan teks yang ada didalam artikel

```python
def clean_text(text):
    return re.sub(r"[^a-zA-Z\s]", "", text)
```

### Membaca semua artikel dari Direktori /artikel

```python
def read_articles(folder_path):
    articles = {}
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            with open(os.path.join(folder_path, filename), "r", encoding="utf-8") as file:
                articles[filename] = file.read()
    return articles

folder_path = "./artikel"
articles = read_articles(folder_path)
```

### Menghitung len (banyak) Paragraf

```python
def count_paragraphs(text): 
    return len([p for p in text.split("\n") if p.strip()])
```

### Menghitung len (banyak) Kalimat

```python
def count_sentences(text):
    return len(re.split(r"[.!?](?=\s|$)", text))
```

### Menghitung frekuensi alfabet a-z

```python
def alphabet_frequency(text):
    text = clean_text(text).lower()
    return Counter(filter(str.isalpha, text))
```

### Menghitung frekuensi kata dalam artikel

```python
# Fungsi untuk menghitung frekuensi kata
def word_frequency(text):
    words = clean_text(text).lower().split()
    return Counter(words)
```

### Menghitung jumlah kata yang bersifat unique ( Unik )

```python
# Fungsi untuk menghitung jumlah kata unik
def unique_words_count(text):
    words = set(clean_text(text).lower().split())
    return len(words)
```

### Memproses setiap artikel dengan perulangan

##### Kernel ini memanggil fungsi : 
1. count_paraghraps ( Menghitung paragraf )
2. count_sentences ( Menghitung banyak paragraf )
3. word_frequency ( Menghitung frekuensi kata dalam artikel )
4. alphabet_frequency ( Menghitung frekuensi alfabet a - z )
5. unique_words ( Menghitung banyak kata yang bersifat unik )

```python
# Proses setiap artikel
for filename, article in articles.items():
    paragraphs = count_paragraphs(article)
    sentences = count_sentences(article)
    length = len(article)
    word_freq = word_frequency(article)
    alphabet_freq = alphabet_frequency(article)
    unique_words = unique_words_count(article)

    print(f"Artikel: {filename}")
    print(f"Jumlah Paragraf: {paragraphs}")
    print(f"Jumlah Kalimat: {sentences}")
    print(f"Panjang Teks: {length}")
    print(f"Jumlah Kata Berbeda: {unique_words}")
    print("Frekuensi Alfabet:")
    print(alphabet_freq)

    # Plot frekuensi alfabet
    plt.figure(figsize=(10, 5))
    plt.bar(alphabet_freq.keys(), alphabet_freq.values(), color="skyblue")
    plt.title(f"Frekuensi Alfabet (A-Z) - {filename}")
    plt.xlabel("Alfabet")
    plt.ylabel("Frekuensi")
    plt.show()

    # Plot frekuensi kata (10 kata paling sering)
    most_common_words = word_freq.most_common(10)
    words, frequencies = zip(*most_common_words)

    plt.figure(figsize=(10, 5))
    plt.bar(words, frequencies, color="lightgreen")
    plt.title(f"10 Kata Paling Sering Muncul - {filename}")
    plt.xlabel("Kata")
    plt.ylabel("Frekuensi")
    plt.show()

    # Grafik garis frekuensi alfabet
    plt.figure(figsize=(10, 5))
    plt.plot(list(alphabet_freq.keys()), list(alphabet_freq.values()), marker="o", color="orange")
    plt.title(f"Frekuensi Alfabet (A-Z) - {filename}")
    plt.xlabel("Alfabet")
    plt.ylabel("Frekuensi")
    plt.grid(True)
    plt.show()
```
